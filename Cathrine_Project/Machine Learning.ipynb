```python
# Cleaning the dataset

# Convert 'Null' strings to actual NaN values
df.replace("Null", pd.NA, inplace=True)

# Convert numeric columns stored as strings to appropriate numeric types
numeric_columns = [
    "Final Score", "Final Practical", "Final Theory", "Final Theory.1",
    "Practical(A) out of 80", "Oral Exam out of 20", "Final Practical.1",
    "Final Icare (Theory + Practical)", "Final Icare (Theory + Practical).1",
    "Hospital Internship Score", "Final Score.1", "CAT 1", "CAT 2", "CAT 3", 
    "CAT 4", "CAT 5", "CAT 6", "Age"
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop duplicate columns if they contain identical data
if df["Final Theory"].equals(df["Final Theory.1"]):
    df.drop(columns=["Final Theory.1"], inplace=True)

if df["Final Practical"].equals(df["Final Practical.1"]):
    df.drop(columns=["Final Practical.1"], inplace=True)

if df["Final Score"].equals(df["Final Score.1"]):
    df.drop(columns=["Final Score.1"], inplace=True)

# Standardize column names by removing spaces and special characters
df.columns = df.columns.str.strip().str.replace(r'[^\w]', '_', regex=True)

# Ensure 'Tel_no' is standardized to start with '+254'
df['Tel_no'] = df['Tel_no'].astype(str).str.replace(r'^(?:254|07|7)', '+254', regex=True)
```

```python
import csv

def add_com_suffix(email):
    # Check if the email already ends with .com
    if not email.endswith('.com'):
        return email + '.com'
    return email

#csv path to read from
csv_file = input("Please enter the CSV file name or path: ")

# Open and read the CSV file with emails
with open('New Caregivers - New Caregivers.csv', 'r') as file:
    reader = csv.reader(file)
  
    next(reader, None)  # This skips the first row (header)
    
    # Extract emails from the CSV
    emails = [row[4].strip() for row in reader]  # Assumes emails are in the first column

# Process the emails to add .com where necessary
valid_emails = [add_com_suffix(email) for email in emails]


print("Processing complete! Valid emails saved to 'valid_emails.csv'.")

# Save the valid emails into a new CSV file
with open('New_Caregivers_updated.csv', 'w', newline='') as output_file:
    writer = csv.writer(output_file)
```

```python
# Optionally, write the valid emails to a new CSV file
with open('valid_emails.csv', 'w', newline='') as output_file:
    writer = csv.writer(output_file)
    # Write header (optional)
    writer.writerow(['Email'])
    
    # Write the processed valid emails
    for email in valid_emails:
        writer.writerow([email])
```
